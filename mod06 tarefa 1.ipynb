#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv("b1abb33c4f.csv")
ap = df[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
ap
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
ap['APGAR1'].value_counts().sort_index().plot.bar(title='Distribuição de APGAR1')
plt.show()
ap['APGAR5'].value_counts().sort_index().plot.bar(title='Distribuição de APGAR5')
plt.show()
ap['SEXO'].value_counts().sort_index().plot.bar(title='Distribuição de SEXO')
plt.show()
ap['GESTACAO'].value_counts().sort_index().plot.bar(title='Distribuição de GESTACAO')
plt.show()
ap['ESTCIVMAE'].value_counts().sort_index().plot.bar(title='Distribuição de ESTCIVMAE')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5', y='CONSULTAS', data=df)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(23, 18))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=ap,)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=ap)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=ap)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df)
plt.show()
```
